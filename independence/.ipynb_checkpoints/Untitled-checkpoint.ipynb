```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
dataset_path = "Colony1-day01.txt"
```

```python
G = nx.read_edgelist(dataset_path, data=[("weight", int)])
```

```python
degree_centrality = nx.degree_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
```

```python
# Set a threshold for low degree centrality and closeness centrality (adjust as needed)
degree_threshold =0.4
closeness_threshold = 0.6
```

```python
# Identify ants with low degree centrality
ants_low_degree = [node for node, centrality in degree_centrality.items() if centrality < degree_threshold]
```

```python
# Identify ants with low closeness centrality
ants_low_closeness = [node for node, centrality in closeness_centrality.items() if centrality < closeness_threshold]
```

```python
# Visualize the graph with nodes having low degree and closeness centrality highlighted
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=500, node_color='skyblue')
```

```python
# Highlight nodes with low degree centrality
nx.draw_networkx_nodes(G, pos, nodelist=ants_low_degree, node_color='red', node_size=500)
```

```python
# Highlight nodes with low closeness centrality
nx.draw_networkx_nodes(G, pos, nodelist=ants_low_closeness, node_color='green', node_size=500)
```

```python
plt.title('Graph with Nodes of Low Degree and Closeness Centrality Highlighted')
plt.show()
```

```python
# Print the identified ants with low degree and closeness centrality
print("Ants with Low Degree Centrality:", ants_low_degree)
print("Ants with Low Closeness Centrality:", ants_low_closeness)
```
